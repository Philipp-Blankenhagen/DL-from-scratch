```python
x_1=np.array(df[df['y']==0].iloc[:,0:2]).T
x_2=np.array(df[df['y']==1].iloc[:,0:2]).T
X = np.array(df.iloc[:,0:2]).T
```

```python
fig = plt.figure(figsize=(20,20))
ax = fig.add_subplot(2,2,1)
ax2 = fig.add_subplot(2,2,2)
ax3 = fig.add_subplot(2,2,3)


f = np.zeros(19)
pct = np.zeros(19)


for cnt,i in enumerate(range_v):
    v = np.array([np.cos(np.pi * i/180),np.sin(np.pi * i/180)])
    ax.plot((0,v[0]),(0,v[1]))
    # projection
    f = (v[None,:]@X >=0) *1
    pct[cnt] = sum(df['y']==f[0,:])/200

ax.scatter(df[df['y']==0].iloc[:,0],df[df['y']==0].iloc[:,1],color='red',label='x.1')
ax.scatter(df[df['y']==1].iloc[:,0],df[df['y']==1].iloc[:,1],color='blue',label='x.2')
ax.legend()
ax.set_xlim((-2,2))
ax.set_ylim((-2,2))

ax2.plot(range_v,pct)
ax2.set_xlabel('Angle')
ax2.set_ylabel('pct correct classifications')


ax3.scatter(df[df['y']==0].iloc[:,0],df[df['y']==0].iloc[:,1],color='red',label='x.1')
ax3.scatter(df[df['y']==1].iloc[:,0],df[df['y']==1].iloc[:,1],color='blue',label='x.2')
ax3.legend()
ax3.set_xlim((-2,2))
ax3.set_ylim((-2,2))



plt.grid(True,'both')

```
