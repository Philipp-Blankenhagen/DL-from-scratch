```python
class MLP:
    'MLP'
    def __init__(self):
        self.w_21 = np.random.normal(0,1,size=(10,1))
        self.w_10_1 = np.random.normal(0,2,size=(1))[0]
        self.w_10_0 = np.random.uniform(-2,2,size=(1))[0]
        print('initialized weights')
        
    def output(self,x):
        return np.sum(self.w_21 * np.tanh(self.w_10_1*x-self.w_10_0))

def MSE(x,y):
    return (1/x.shape[0]) * sum((x - y)**2)
    
def output(w_21,w_10_1,w_10_0,x):
    return np.sum(w_21 * np.tanh(w_10_1*x-w_10_0))
```

```python
models ={}

for i in range(0,50):
    models[i] = MLP()
    models[f'{i}_y'] = np.zeros((100,2))
    for cnt, x in enumerate(np.arange(-2,2,1/25)):
        models[f'{i}_y'][cnt,0] = x
        models[f'{i}_y'][cnt,1] = models[i].output(x)
```

```python
def d_tanh(value):
    return 1 - np.tanh(value)**2
```

```python
class backprop():
    def __init__(self):
        if True:
            self.w_L_1 = np.random.uniform(-0.5,0.5,size=(4,1))
            self.w_1_0 = np.random.uniform(-0.5,0.5,size=(3,2))
        if False:
            self.w_L_1 = np.ones((4,1))*0.5
            self.w_1_0 = np.ones((3,2))*0.5
        print('initialization')
        
    def forward_pass(self,x):
        
        #  Input Layer
        self.x = x
        x_bias = np.array([1,x])[:,None] # x values with bias
        self.h_0 = x_bias
        self.s_0 = self.h_0
        
        
        #  Hidden Layer
        #print('s_0:',self.s_0,'\n\nw_1_0:',self.w_1_0)
        self.h_1 = self.w_1_0@self.s_0  # Input for hidden layer
        self.h_1_bias = np.insert(self.h_1, 0, values=1)[:,None]
        self.s_1_weight = np.tanh(self.h_1) # s_1_weight doesnt include bias
        self.s_1= np.insert(self.s_1_weight, 0,values=1)[:,None] # add bias to output of hidden layer
        
        #print('h_1:',self.h_1)
        #print('s_1_weight',self.s_1_weight)
        
        #  Output Layer
        self.h_L = self.w_L_1.T@self.s_1
        self.y = np.tanh(self.h_L)[0][0]
        
        #print('w_L_1',self.w_L_1,'h_L',self.h_L)
        
        return self.y
    
    
    def backward_pass(self):
        self.d_L = d_tanh(self.h_L)
        self.d_1_v = d_tanh(self.h_1_bias) * (self.d_L * self.w_L_1) 
        self.d_0_0 = d_tanh(1) * sum(self.d_1_v[1:]*self.w_1_0[:,0][:,None]) # Bias
        self.d_0_1 = d_tanh(self.x) * sum(self.d_1_v[1:]*self.w_1_0[:,1][:,None]) # Weight
        
        
        #print('h_1_weight:',self.h_1_weight)
        #print('d_L:',self.d_L)
        #print('d_1_v:',self.d_1_v)
        #print('d_0_0:',self.d_0_0)
        #print('d_0_1:',self.d_0_1)
        #print('x',self.x)
        
        return
    def gradient(self):
        self.grad_w_L_1 += self.error * self.d_L * self.s_1
        self.grad_w_1_0 += self.error * self.d_1_v[1:] @ self.s_0.T
        return
        
    def weight_update(self,eta,n):
        #print(y,self.y)
        self.w_L_1 = self.w_L_1 - (eta * (1/n) * self.grad_w_L_1)
        self.w_1_0 = self.w_1_0 - (eta * (1/n) * self.grad_w_1_0)
                                   
        #print('weight\n',self.w_L_1)
        # Reset gradient

        #print('error:',self.error)
        #print('')
        return
    
    def fit(self,x,y,n_epoch,eta):
        self.x_all = x
        self.y_true = y
        self.n_epoch = n_epoch
        self.epoch_error = np.ones((n_epoch,1))
        self.d_L = 0
        self.d_1_v = 0
        self.d_0_0 = 0
        self.d_0_1 = 0
        
        for epoch in range(0,n_epoch):
            self.grad_w_L_1 = 0 # reset gradient
            self.grad_w_1_0 = 0 # reset gradient

            self.batch_error = 0
            for sample in range(0,len(x)):
                self.forward_pass(x[sample])
                self.error =  self.y - y[sample]
                self.backward_pass()   
                self.gradient()
                self.batch_error += self.error**2
                #print('output',self.y,y[sample],self.error)
                
            
            self.weight_update(eta,len(x)) # update weights after batch
            self.epoch_error[epoch] = (1/len(x)) * self.batch_error
        
        return
    
    def plotting(self):
        fig = plt.figure(figsize=(20,20))
        ax1 = fig.add_subplot(2,2,1)
        ax2 = fig.add_subplot(2,2,2)
        #ax3 = fig.add_subplot(2,2,3)
        
        y_predictions = [self.forward_pass(i) for i in self.x_all]
        y_tuple = list(zip(model.y_true,y_predictions))
        ax1.plot(np.arange(0,self.n_epoch),self.epoch_error)
        ax2.scatter(self.x_all,self.y_true,label='original dataset',color='blue')
        ax2.scatter(self.x_all,y_predictions,color='red',label='predictions')
        ax2.plot((self.x_all,self.x_all),([i for (i,j) in y_tuple], [j for (i,j) in y_tuple]),c='black')


        ax2.legend()
        plt.show()
```

```python
class MLP_torch_manual(T.nn.Module): # Class needs to be from torch.nn.Module
    def __init__(self):
        super().__init__() 
        
        
        if False: 
            # Add the layers there the variable names are not important
            # Then in forward pass the order will be set
            self.hidden = T.nn.Linear(1,3,bias=True) # hidden Layer
            self.output = T.nn.Linear(3,1,bias=True) # Output Layer
            self.tanh = T.nn.Tanh() #  apply tanh
            
            
            # Initialization with weights
            if True: # random initialization
                T.nn.init.uniform_(self.output.weight,-0.5,0.5)
                T.nn.init.uniform_(self.output.bias,-0.5,0.5)
                T.nn.init.uniform_(self.hidden.weight,-0.5,0.5)
                T.nn.init.uniform_(self.hidden.bias,-0.5,0.5)


            if False: # Weights with same values of 0.5
                T.nn.init.constant_(self.output.weight,0.5)
                T.nn.init.constant_(self.output.bias,0.5)
                T.nn.init.constant_(self.hidden.weight,0.5)
                T.nn.init.constant_(self.hidden.bias,0.5)


            
        if True: # Easy way using sequential
            self.layer = []
            self.layer.append(T.nn.Linear(1,3,bias=True)) # Hidden layer
            self.layer.append(T.nn.Tanh()) # Activation for hidden layer
            self.layer.append(T.nn.Linear(3,1,bias=True)) # Output layer
            self.layer.append(T.nn.Tanh()) # Activation for output layer
            self.main = nn.Sequential(*self.layer)
            
            if False:
                T.nn.init.constant_(self.layer[0].weight,0.5)
                T.nn.init.constant_(self.layer[2].weight,0.5)
                T.nn.init.constant_(self.layer[0].bias,0.5)
                T.nn.init.constant_(self.layer[2].bias,0.5)

            
            
            
    def forward(self, x):
        
        if False:
            h_0 = x
            h_1 = self.hidden(h_0) # without bias
            s_1 = self.tanh(h_1)
            h_L = self.output(s_1) # Output with bias after hidden layer
            y = self.tanh(h_L)
            
        if True:
            y = self.main(x)
        
        return y
    
    
    def fit(self,dl,n_epochs,lr):
        self.n_epoch = n_epochs
        self.dl = dl
        opt = T.optim.SGD(self.parameters(), lr)
        MSE = nn.MSELoss()
        self.losses = np.ones((n_epochs,1))
        
        for epoch in range(0,n_epochs):
            epoch_loss=0
            
            for batch, (X,y) in enumerate(dl):
                print('X shape',X.shape)
                print('y shape',y.shape)
                
                opt.zero_grad()
                pred = self.forward(X)
                loss = MSE(pred, y)
                print('pred shape',pred.shape)
                print('loss shape',loss.shape)
                epoch_loss += loss
                loss.backward()
                opt.step()
                
            #print(epoch_loss)
            self.losses[epoch] = epoch_loss.detach().numpy().item()

            
    def plotting(self):
        self.x_all = self.dl.batch_sampler.sampler.data_source[:][0]
        self.y_true = self.dl.batch_sampler.sampler.data_source[:][1]

        fig = plt.figure(figsize=(20,20))
        ax1 = fig.add_subplot(2,2,1)
        ax2 = fig.add_subplot(2,2,2)
        #ax3 = fig.add_subplot(2,2,3)
        
        
        y_predictions = [self.forward(i[None]).detach().numpy()[0] for i in self.x_all]
        y_tuple = list(zip(self.y_true,y_predictions))
        
        ax1.plot(np.arange(0,self.n_epoch),self.losses)
        ax2.scatter(self.x_all.detach().numpy(),self.y_true.detach().numpy(),label='original dataset',color='blue')
        ax2.scatter(self.x_all.detach().numpy(),y_predictions,color='red',label='predictions')
        ax2.plot((self.x_all.detach().numpy()[:,0],self.x_all.detach().numpy()[:,0]),([i for (i,j) in y_tuple], [j for (i,j) in y_tuple]),c='black')


        ax2.legend()
        plt.show()
```
